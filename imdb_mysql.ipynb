```python
import MySQLdb as mdb

con = mdb.connect('localhost', 'root', '', 'imdb');
cur = con.cursor()

# cur.execute("SELECT VERSION()")
# ver = cur.fetchone()
# print "Database version : %s " % ver
```

```python
cur.execute('CREATE OR REPLACE VIEW top_voted_movies AS SELECT movie_id, title, production_year AS year, info AS votes FROM title \
JOIN movie_info_idx ON title.id = movie_id WHERE info_type_id = 100 AND CONVERT(movie_info_idx.info, UNSIGNED INTEGER) > 10000 \
AND title.kind_id = 1;')
```

```python
cur.execute("CREATE OR REPLACE VIEW movie_genre AS SELECT movie_id, info_type_id, info FROM movie_info WHERE info_type_id = 3 \
            AND movie_id IN (SELECT movie_id FROM top_voted_movies);")
```

```python
cur = con.cursor()
cur.execute('SELECT * FROM top_voted_movies;')

top_voted = DataFrame(list(cur.fetchall()), columns = ['movie_id', 'title', 'year', 'votes'])
top_voted[:10]
```

```python
# to get urls from search with name and year
ia = imdb.IMDb()
top_voted['url'] = 'empty'
tic = timeit.default_timer()

for i in range(top_voted.shape[0]):
# for i in range(10):
    movie = top_voted.iloc[i]['title'] + ' ' + str(top_voted.iloc[i]['year'])
    results = ia.search_movie(movie)
    if results:
        result = results[0]
        top_voted.ix[i,'url'] = ia.get_imdbURL(result)
timeit.default_timer() - tic
top_voted.to_csv('top_voted_url.txt', sep='\t', index = False)
```
